```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
root_dir = '../'
distance_files = []
for root, dirs, files in os.walk(root_dir):
    for file in files:
        if file == 'distance.csv':
            distance_files.append(os.path.join(root, file))

distance_files
```

```python
# [distance_files[i] for i in [2,6,8,14,15]]
distance_files = [distance_files[i] for i in [2,15,8,6,14]]
```

```python
out = pd.DataFrame()
for d in distance_files:
    df = pd.read_csv(d)
    df_grouped = df.groupby('shape')
    for key in df_grouped.groups.keys():
        df_sub = df_grouped.get_group(key).reset_index(drop=True)
        v_mean = df_sub.loc[:, 'v_rms']*1e4
        colname = Path(d).parents[1].name+'_'+key
        out.loc[:, colname] = v_mean
        # break
```

```python
out.to_csv('average_speed_rms.csv')
```
